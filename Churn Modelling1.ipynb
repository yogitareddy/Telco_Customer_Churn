# 1. Imported Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 3. Identified any int or float values which are used as an object and converted them to float or integer.

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

# 4. Identified the missing values and imputed them.

```python
df.isnull().sum()
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
df['TotalCharges'].isnull().sum()
```

# 5. Replaced no internet service with 'No'.

```python
df=df.replace({'No internet service': 'No'})
df.head()
```

```python
for i in df.columns:
    print(i)
    print(df[i].unique())
```

# 6. Replaced no phone service with 'No'.

```python
df=df.replace({'No phone service': 'No'})
```

```python
for i in df.columns:
    print(i,':')
    print(df[i].unique())
```

# 7. Encoded the categorical values to binary integers.

```python
cols=['gender','Partner','Dependents','PhoneService','PaperlessBilling','Churn','OnlineSecurity','OnlineBackup',
      'DeviceProtection','TechSupport','StreamingMovies','StreamingTV','MultipleLines']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cols:
    df[col]=le.fit_transform(df[col])
df.head(5)
```

# 8. Used One Hot Encoding on more than 2 categorical values.

```python
pd.get_dummies(df.Contract)
```

```python
pd.get_dummies(df.PaymentMethod)
```

```python
pd.get_dummies(df.InternetService)
```

```python
df = pd.concat([df.drop('Contract',axis=1), pd.get_dummies(df['Contract'])], axis=1)
df = pd.concat([df.drop('PaymentMethod',axis=1), pd.get_dummies(df['PaymentMethod'])], axis=1)
df = pd.concat([df.drop('InternetService',axis=1), pd.get_dummies(df['InternetService'])], axis=1)
```

# 12. Divide data into X and y. then train test split.

```python
X=df.drop(columns=['Churn'],axis=1)
y=df['Churn']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

# 13. Implemented Logistic regression, Decision Tree Classifier, Random Forrest Classifier.

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model=LogisticRegression(solver='liblinear')
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred=lr_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

# Classification Report for each model

```python
from sklearn.metrics import classification_report
```

```python
target_names = ['class 0', 'class 1']
print(classification_report(y_test, y_pred,target_names=target_names))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC_model=RandomForestClassifier()
```

```python
RFC_model.fit(x_train,y_train)
```

```python
y_pred=RFC_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
